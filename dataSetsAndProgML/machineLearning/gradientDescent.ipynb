```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradientDescent(xPoints,yPoints):
    mOld=0
    cOld =0
    learningRate = 0.1
    n=15
    iterations = 1000
    
    for i in range(iterations):
        slopeDerivative = -(1/n)*(yPoints - mOld*xPoints - cOld)*xPoints
        interceptDerivative = -(1/n)*(yPoints - mOld*xPoints - cOld)
    
        mNew = mOld - learningRate*slopeDerivative
        cNew = cOld - learningRate*interceptDerivative
    
    plt.scatter(mNew,cNew)
    plt.plot(mNew,cNew,color = "g") 
    plt.show() 
```

```python
xPoints=np.array([6,5,8,10,12,3,35,6,61,25,34,21,41,53,51])
yPoints=np.array([2,6,3,4,21,23,25,52,51,12,43,12,35,10,11])

gradientDescent(xPoints,yPoints)
```
